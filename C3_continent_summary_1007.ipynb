```python
import pandas as pd
import numpy as numpy
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
combi_data = pd.read_csv('post_WWII_data_1002.csv')
combi_data.head()
```

```python
grouped_continent_home_sum = combi_data.groupby(by=['home_continent']).sum()
grouped_continent_home_sum.head()
```

```python
continent_home_results = pd.DataFrame(combi_data.groupby(by=combi_data['home_continent']).home_result.value_counts())
continent_home_results
```

```python
continent_away_results = pd.DataFrame(combi_data.groupby(by=combi_data['away_continent']).home_result.value_counts())
continent_away_results
```

```python
cleaned_home_df = grouped_continent_home_sum.drop(columns=['Unnamed: 0','neutral','year','month','lat','lng','result'])
cleaned_home_df['ttl_games'] = ''
cleaned_home_df['home_games']=''
cleaned_home_df['away_games']=''
cleaned_home_df['ttl_wins'] = ''
cleaned_home_df['home_wins'] = ''
cleaned_home_df['away_wins'] = ''
cleaned_home_df
```

```python
# summarizing the total games.
cleaned_home_df.loc['Africa','home_games']=continent_home_results.loc['Africa','home_result']['win']+\
continent_home_results.loc['Africa','home_result']['lose']+continent_home_results.loc['Africa','home_result']['Tie']

cleaned_home_df.loc['Asia','home_games']=continent_home_results.loc['Asia','home_result']['win']+\
continent_home_results.loc['Asia','home_result']['lose']+continent_home_results.loc['Asia','home_result']['Tie']

cleaned_home_df.loc['Europe','home_games']=continent_home_results.loc['Europe','home_result']['win']+\
continent_home_results.loc['Europe','home_result']['lose']+continent_home_results.loc['Europe','home_result']['Tie']

cleaned_home_df.loc['NCAC','home_games']=continent_home_results.loc['NCAC','home_result']['win']+\
continent_home_results.loc['NCAC','home_result']['lose']+continent_home_results.loc['NCAC','home_result']['Tie']

cleaned_home_df.loc['Oceania','home_games']=continent_home_results.loc['Oceania','home_result']['win']+\
continent_home_results.loc['Oceania','home_result']['lose']+continent_home_results.loc['Oceania','home_result']['Tie']

cleaned_home_df.loc['South America','home_games']=continent_home_results.loc['South America','home_result']['win']+\
continent_home_results.loc['South America','home_result']['lose']+continent_home_results.loc['South America','home_result']['Tie']

cleaned_home_df
```

```python
cleaned_home_df.loc['Africa','home_wins']=continent_home_results.loc['Africa','home_result']['win']
cleaned_home_df.loc['Asia','home_wins']=continent_home_results.loc['Asia','home_result']['win']
cleaned_home_df.loc['Europe','home_wins']=continent_home_results.loc['Europe','home_result']['win']
cleaned_home_df.loc['NCAC','home_wins']=continent_home_results.loc['NCAC','home_result']['win']
cleaned_home_df.loc['Oceania','home_wins']=continent_home_results.loc['Oceania','home_result']['win']
cleaned_home_df.loc['South America','home_wins']=continent_home_results.loc['South America','home_result']['win']

cleaned_home_df
```

```python
grouped_continent_away_sum = combi_data.groupby(by=['away_continent']).sum()
grouped_continent_away_sum.head()
```

```python
cleaned_away_df = grouped_continent_away_sum.drop(columns=['Unnamed: 0','neutral','year','month','lat','lng','result'])
cleaned_away_df['ttl_games'] = ''
cleaned_away_df['home_games']=''
cleaned_away_df['away_games']=''
cleaned_away_df['ttl_wins'] = ''
cleaned_away_df['home_wins'] = ''
cleaned_away_df['away_wins'] = ''
cleaned_away_df=cleaned_away_df.drop(['Others'],axis=0)
cleaned_away_df
```

```python
cleaned_away_df.loc['Africa','away_games']=continent_away_results.loc['Africa','home_result']['win']+\
continent_away_results.loc['Africa','home_result']['lose']+continent_away_results.loc['Africa','home_result']['Tie']

cleaned_away_df.loc['Asia','away_games']=continent_away_results.loc['Asia','home_result']['win']+\
continent_away_results.loc['Asia','home_result']['lose']+continent_away_results.loc['Asia','home_result']['Tie']

cleaned_away_df.loc['Europe','away_games']=continent_away_results.loc['Europe','home_result']['win']+\
continent_away_results.loc['Europe','home_result']['lose']+continent_away_results.loc['Europe','home_result']['Tie']

cleaned_away_df.loc['NCAC','away_games']=continent_away_results.loc['NCAC','home_result']['win']+\
continent_away_results.loc['NCAC','home_result']['lose']+continent_away_results.loc['NCAC','home_result']['Tie']

cleaned_away_df.loc['Oceania','away_games']=continent_home_results.loc['Oceania','home_result']['win']+\
continent_away_results.loc['Oceania','home_result']['lose']+continent_away_results.loc['Oceania','home_result']['Tie']

cleaned_away_df.loc['South America','away_games']=continent_home_results.loc['South America','home_result']['win']+\
continent_away_results.loc['South America','home_result']['lose']+continent_away_results.loc['South America','home_result']['Tie']

cleaned_away_df
```

```python
cleaned_away_df.loc['Africa','away_wins']=continent_away_results.loc['Africa','home_result']['lose']

cleaned_away_df.loc['Asia','away_wins']=continent_away_results.loc['Asia','home_result']['lose']

cleaned_away_df.loc['Europe','away_wins']=continent_away_results.loc['Europe','home_result']['lose']

cleaned_away_df.loc['NCAC','away_wins']=continent_away_results.loc['NCAC','home_result']['lose']

cleaned_away_df.loc['Oceania','away_wins']=continent_away_results.loc['Oceania','home_result']['lose']

cleaned_away_df.loc['South America','away_wins']=continent_away_results.loc['South America','home_result']['lose']

cleaned_away_df
```

```python
cleaned_combined_df = cleaned_home_df
cleaned_combined_df
```

```python
cleaned_combined_df['home_lost_goals']=cleaned_combined_df['away_score']
cleaned_combined_df['away_goals']=cleaned_away_df['away_score']
cleaned_combined_df['away_wins']=cleaned_away_df['away_wins']
cleaned_combined_df['away_games']=cleaned_away_df['away_games']
cleaned_combined_df['ttl_games']=cleaned_combined_df['home_games']+cleaned_combined_df['away_games']
cleaned_combined_df['ttl_wins']=cleaned_combined_df['home_wins']+cleaned_combined_df['away_wins']
cleaned_combined_df.drop(columns=['away_score'])
cleaned_combined_df['away_lost_goals']=cleaned_away_df['home_score']
cleaned_combined_df
```

```python
cleaned_combined_df['winning rate']=cleaned_combined_df['ttl_wins']/cleaned_combined_df['ttl_games']
cleaned_combined_df['ttl_goals']=cleaned_combined_df['home_score'] + cleaned_combined_df['away_goals']
cleaned_combined_df['ttl_lost_goals']=cleaned_combined_df['home_lost_goals'] + cleaned_combined_df['away_lost_goals']
cleaned_combined_df['net_goals']=cleaned_combined_df['ttl_goals'] - cleaned_combined_df['ttl_lost_goals']
cleaned_combined_df
```

```python
cleaned_combined_df.to_csv('continent_overview_pwii_1007.csv')
```

```python
combined_continent = cleaned_combined_df.reset_index()
combined_continent_ranked = combined_continent.sort_values('winning rate',ascending=False)
```

```python
fig,ax1 = plt.subplots()
 
ax1.bar(combined_continent_ranked['home_continent'],combined_continent_ranked['winning rate'],color='cornflowerblue',align='center',width=0.5)
ax1.tick_params(axis='y', labelcolor='cornflowerblue')
ax1.tick_params(axis='x', labelrotation=-90,labelsize=10)
ax1.set_ylim(0,0.5)
ax1.set_title('Continent Winning Rate vs Net Goals')
ax1.legend('Win Rate',loc=2,fontsize=10,handlelength = 3,frameon=False)

ax2 = ax1.twinx()
ax2.plot(combined_continent_ranked['home_continent'],combined_continent_ranked['net_goals'],color='darkgreen')
ax2.tick_params(axis='y', labelcolor='darkgreen')
ax2.legend('Net Goals',loc=9,fontsize=10,frameon=False)
ax2.hlines(y=0,xmin=-.25,xmax=5.25,color='r',linestyle='--')
```
