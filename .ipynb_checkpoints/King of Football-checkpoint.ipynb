```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import calendar
```

```python
csv_path = 'results.csv'
master_data = pd.read_csv(csv_path)
master_df = pd.DataFrame(master_data)
master_df.info()
```

```python
master_df['year']=''
master_df['month']=''
master_df['goal_diff']=''
master_df['home_result']=''
master_df['lat']=''
master_df['lng']=''
master_df['host_continent']=''
master_df['home_continent']=''
master_df['away_continent']=''
master_df.head()
```

```python
master_df['year'] = pd.DatetimeIndex(master_df['date']).year
master_df['month'] = pd.DatetimeIndex(master_df['date']).month
master_df['goal_diff']= master_df['home_score'] - master_df['away_score']

master_df.head()
```

```python
master_df['goal_diff']= master_df['goal_diff'].astype('str')
master_df['goal_diff']= master_df['goal_diff'].str.replace('-','')
master_df.head()
```

```python
master_df['home_result'] = np.where(master_df.home_score == master_df.away_score, 'Tie', 
                                 np.where(master_df.home_score > master_df.away_score, 'win', 'lose'))
master_df.head()
```
