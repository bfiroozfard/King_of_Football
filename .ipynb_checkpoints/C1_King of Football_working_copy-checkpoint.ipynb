```python
host_ctry_list = corr_df['country'].unique()
len(host_ctry_list)
```

```python
for index, row in corr_df.iterrows():
    if row['home_team'] in eu_ctry_list:
        corr_df.loc[index,'home_continent'] = 'Europe'
    elif row['home_team'] in ncac_ctry_list:
        corr_df.loc[index,'home_continent'] = 'NCAC'
    elif row['home_team'] in sa_ctry_list:
        corr_df.loc[index,'home_continent'] = 'South America'
    elif row['home_team'] in af_ctry_list:
        corr_df.loc[index,'home_continent'] = 'Africa'
    elif row['home_team'] in as_ctry_list:
        corr_df.loc[index,'home_continent'] = 'Asia'
    elif row['home_team'] in oc_ctry_list:
        corr_df.loc[index,'home_continent'] = 'Oceania'
    else:
        corr_df.loc[index,'home_continent'] = 'Others'
    
    if row['away_team'] in eu_ctry_list:
        corr_df.loc[index,'away_continent'] = 'Europe'
    elif row['away_team'] in ncac_ctry_list:
        corr_df.loc[index,'away_continent'] = 'NCAC'
    elif row['away_team'] in sa_ctry_list:
        corr_df.loc[index,'away_continent'] = 'South America'
    elif row['away_team'] in af_ctry_list:
        corr_df.loc[index,'away_continent'] = 'Africa'
    elif row['away_team'] in as_ctry_list:
        corr_df.loc[index,'away_continent'] = 'Asia'
    elif row['away_team'] in oc_ctry_list:
        corr_df.loc[index,'away_continent'] = 'Oceania'
    else:
        corr_df.loc[index,'away_continent'] = 'Others'
    
    if row['country'] in eu_ctry_list:
        corr_df.loc[index,'host_continent'] = 'Europe'
    elif row['country'] in ncac_ctry_list:
        corr_df.loc[index,'host_continent'] = 'NCAC'
    elif row['country'] in sa_ctry_list:
        corr_df.loc[index,'host_continent'] = 'South America'
    elif row['country'] in af_ctry_list:
        corr_df.loc[index,'host_continent'] = 'Africa'
    elif row['country'] in as_ctry_list:
        corr_df.loc[index,'host_continent'] = 'Asia'
    elif row['country'] in oc_ctry_list:
        corr_df.loc[index,'host_continent'] = 'Oceania'
    else:
        corr_df.loc[index,'host_continent'] = 'Others'
```

```python
corr_df.head()
```

```python
corr_df['year'] = pd.DatetimeIndex(corr_df['date']).year
corr_df['month'] = pd.DatetimeIndex(corr_df['date']).month
corr_df['goal_diff']= corr_df['home_score'] - corr_df['away_score']

corr_df.head()
```

```python
master_df['goal_diff']= master_df['goal_diff'].astype('str')
master_df['goal_diff']= master_df['goal_diff'].str.replace('-','')
master_df.head()
```

```python
master_df['home_result'] = np.where(master_df.home_score == master_df.away_score, 'Tie', 
                                 np.where(master_df.home_score > master_df.away_score, 'win', 'lose'))
master_df.head()
```

```python
corr_df['result']=''
corr_df['total_goals']=''
corr_df['total_goals']=corr_df['home_score']+corr_df['away_score']
corr_df
```

```python
def result(row):
      
    if row['home_score'] > row['away_score']:
        return row['home_team']
    elif row['home_score'] < row['away_score']:
        return row['away_team']
    else:
        return('Tie')
corr_df['results'] = corr_df[['home_score','away_score','home_team','away_team']].apply(result,axis=1)
corr_df.head()
```
