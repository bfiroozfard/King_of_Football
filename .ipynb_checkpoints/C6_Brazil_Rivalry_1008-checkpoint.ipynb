```python
import pandas as pd
import numpy as numpy
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
pwii_combined_data = pd.read_csv('output/post_WWII_data_1002.csv').drop(columns='result')
pwii_combined_data.head()
```

```python
brazil_home_raw=pwii_combined_data.loc[pwii_combined_data.home_team=='Brazil',:]
brazil_home_raw
```

```python
brazil_home_df=brazil_home_raw.groupby(by='away_team').sum().drop(columns=['neutral','year','month','lat','lng']).sort_values('Unnamed: 0',ascending=False)
brazil_home_df.head()
```

```python
brazil_home_rivalry_1 = brazil_home_raw.groupby(by='away_team').count()
brazil_home_rivalry_1 = brazil_home_rivalry_1.sort_values('date',ascending=False)
brazil_home_rivalry = pd.DataFrame(brazil_home_rivalry_1['home_team'])
brazil_home_rivalry['win_counts']=''
brazil_home_rivalry['tie_counts']=''
brazil_home_rivalry['lose_counts']=''
brazil_home_rivalry['win_rate']=''
brazil_home_rivalry
```

```python
away_team_list=list(brazil_home_rivalry.index)
```

```python
brazil_home_result_df = pd.DataFrame(brazil_home_raw.groupby(by='away_team')['home_result'].value_counts())
brazil_home_result_df=brazil_home_result_df.rename(columns={'home_result':'counts'})
brazil_home_result_df
```

```python
for country in away_team_list:
    try:
        brazil_home_rivalry.loc[country,'win_counts']=brazil_home_result_df.loc[country,'win']['counts'].astype(int)
    except (KeyError):
        pass

    try:
        brazil_home_rivalry.loc[country,'tie_counts']=brazil_home_result_df.loc[country,'Tie']['counts'].astype(int)
    except (KeyError):
        pass

    try:
        brazil_home_rivalry.loc[country,'lose_counts']=brazil_home_result_df.loc[country,'lose']['counts'].astype(int)
    except (KeyError):
        pass

brazil_home_rivalry
```

```python
for index,row in brazil_home_rivalry.iterrows():
    try:
        brazil_home_rivalry.loc[index,'win_rate']=int(brazil_home_rivalry.loc[index,'win_counts'])/brazil_home_rivalry.loc[index,'home_team']
    except:
        brazil_home_rivalry.loc[index,'win_rate']=0

brazil_home_rivalry
```

```python
brazil_home_rivalry_reset=brazil_home_rivalry.reset_index()
brazil_home_rivalry_plot = brazil_home_rivalry_reset.iloc[:20,:]
brazil_home_rivalry_plot
```

```python
plt.figsize=(30,15)
plt.bar(brazil_home_rivalry_plot['away_team'],brazil_home_rivalry_plot['win_rate'])
plt.xticks(rotation=90)
plt.title('Brazil Winning Rates At Home',fontsize=15)
plt.hlines(0.654066,xmin=-.5,xmax=20,color='r')
```

```python
brazil_away_raw=pwii_combined_data.loc[pwii_combined_data.away_team=='Brazil',:]
brazil_away_raw
```

```python
brazil_away_df=brazil_away_raw.groupby(by='home_team').sum().drop(columns=['neutral','year','month','lat','lng']).sort_values('Unnamed: 0',ascending=False)
brazil_away_df.head()
```

```python
brazil_away_rivalry_1 = brazil_away_raw.groupby(by='home_team').count()
brazil_away_rivalry_1 = brazil_away_rivalry_1.sort_values('date',ascending=False)
brazil_away_rivalry_1
brazil_away_rivalry = pd.DataFrame(brazil_away_rivalry_1['away_team'])
brazil_away_rivalry['win_counts']=''
brazil_away_rivalry['tie_counts']=''
brazil_away_rivalry['lose_counts']=''
brazil_away_rivalry['win_rate']=''
brazil_away_rivalry
```

```python
home_team_list=list(brazil_away_rivalry.index)
```

```python
brazil_away_result_df = pd.DataFrame(brazil_away_raw.groupby(by='home_team').home_result.value_counts())
brazil_away_result_df=brazil_away_result_df.rename(columns={'home_result':'counts'})
brazil_away_result_df
```

```python
for country in home_team_list:
    try:
        brazil_away_rivalry.loc[country,'win_counts']=brazil_away_result_df.loc[country,'win']['counts']
    except (KeyError):
        pass

    try:
        brazil_away_rivalry.loc[country,'tie_counts']=brazil_away_result_df.loc[country,'Tie']['counts']
    except (KeyError):
        pass

    try:
        brazil_away_rivalry.loc[country,'lose_counts']=brazil_away_result_df.loc[country,'lose']['counts']
    except (KeyError):
        pass

brazil_away_rivalry
```

```python
for index,row in brazil_away_rivalry.iterrows():
    try:
        brazil_away_rivalry.loc[index,'win_rate']=int(brazil_away_rivalry.loc[index,'win_counts'])/brazil_away_rivalry.loc[index,'away_team']
    except:
        brazil_away_rivalry.loc[index,'win_rate']=0

brazil_away_rivalry
```

```python
brazil_away_rivalry_reset=brazil_away_rivalry.reset_index()
brazil_away_rivalry_plot = brazil_away_rivalry_reset.iloc[:20,:]
brazil_away_rivalry_plot
```

```python
plt.figsize=(30,15)
plt.bar(brazil_away_rivalry_plot['home_team'],brazil_away_rivalry_plot['win_rate'])
plt.xticks(rotation=90)
plt.title('Brazil Winning Rates While Away',fontsize=15)
plt.hlines(0.654066,xmin=-.5,xmax=20,color='r')
```
